```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
raw_data=pd.read_csv('1.03. Dummies.csv')
raw_data
```

```python
data=raw_data.copy()
data['Attendance']=data['Attendance'].map({'Yes':1,'No':0})
data
```

```python
data.describe()
```

# Regression

```python
y=data['GPA']
x1=data[['SAT','Attendance']]
```

```python
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data['SAT'],y)
yhat_no=0.6439+0.0014*data['SAT']
yhat_yes=0.8665+0.0014*data['SAT']
fig=plt.plot(data['SAT'],yhat_no,lw=2,c='#CD5C5C')
fig=plt.plot(data['SAT'],yhat_yes,lw=2,c='#FFA07A')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```

So,we have considered attendance as a dummy variable here.

If the candidate has covered the minimum 75 percent of attendance, what's going to be effect on the gpa is to be determined.

From the OLS regression results we found out the coffecients. 

We devided the model into two sub models, i.e one) candidtate has passed the criteria of minimum attendance of 75 percent
and the other one complememtry of that.

so regression is as follows:
 
 GPA=0.6439+0.0014*SAT   -> NO
 
 GPA=0.8665+0.0014*SAT -> YES

# How to make predictions based on the regression we create

Now we will make the prediction of the GPA of two students named:

Student A who scored 1700 in SAT but did not attended the 75 percent of lectures.

Student B who scored 1670 in SAT, moreover attended 75 percent of lectures.

```python
new_data=pd.DataFrame({'const':1,'SAT':[1700,1670],'Attendance':[0,1]})
new_data=new_data[['const','SAT','Attendance']]
new_data
```

```python
new_data.rename(index={0:'A',1:'B'})
```

```python
prediction=results.predict(new_data)
prediction
```

```python
predictiondf=pd.DataFrame({'Prediction':prediction})
joineddf=new_data.join(predictiondf)
joineddf.rename({0:'A',1:'B'})
```

So our model has predicted the GPA of student A and student B to be 3.023513 and 3.204163 respectively.
